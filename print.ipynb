```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Datos de la matriz de correlaci√≥n
data = {
    'Ltr de gas': [1, 0.8, 0.6, 0.4, 0.8, 0.7],
    'Ltr de prod': [0.8, 1, 0.8, 0.4, 0.6, 0.6],
    'Tiempo': [0.6, 0.8, 1, 0.4, 0.6, 0.6],
    'Temp in': [0.4, 0.4, 0.4, 1, 0.4, 0.4],
    'Temp fin': [0.8, 0.6, 0.6, 0.4, 1, 0.6],
    'Conduct': [0.7, 0.6, 0.6, 0.4, 0.6, 1]
}

# Crear un DataFrame con los datos
df = pd.DataFrame(data, index=['Ltr de gas', 'Ltr de prod', 'Tiempo', 'Temp in', 'Temp fin', 'Conduct'])
                
# Crear el mapa de calor
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(df, annot=True, cmap='coolwarm', vmin=0, vmax=1, cbar_kws={'label': 'Correlaci√≥n'})
plt.title('Matriz de Combinaciones')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Datos de la matriz de correlaci√≥n incluyendo "Geom Base"
data = {
    'Ltr de gas': [1, 0.8, 0.6, 0.4, 0.8, 0.7, 0.5],
    'Ltr de prod': [0.8, 1, 0.8, 0.4, 0.6, 0.6, 0.6],
    'Tiempo': [0.6, 0.8, 1, 0.4, 0.6, 0.6, 0.6],
    'Temp in': [0.4, 0.4, 0.4, 1, 0.4, 0.4, 0.4],
    'Temp fin': [0.8, 0.6, 0.6, 0.4, 1, 0.6, 0.7],
    'Conduct': [0.7, 0.6, 0.6, 0.4, 0.6, 1, 0.8],
    'Geom Base': [0.5, 0.6, 0.6, 0.4, 0.7, 0.8, 1]
}

# Crear un DataFrame con los datos
df = pd.DataFrame(data, index=['Ltr de gas', 'Ltr de prod', 'Tiempo', 'Temp in', 'Temp fin', 'Conduct', 'Geom Base'])

# Crear el mapa de calor
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(df, annot=True, cmap='coolwarm', vmin=0, vmax=1, cbar_kws={'label': 'Correlaci√≥n'})
plt.title('Matriz de Correlaci√≥n')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter


# Create the DataFrame
data = {
    "A√±o": list(range(1, 16)),
    "Precio por litro ($MXN)": [9.95, 11.24, 12.71, 14.36, 16.22, 18.33, 20.72, 23.41, 26.45, 29.89, 33.78, 38.17, 43.13, 48.74, 55.07],
    "Ahorro anual ($MXN)": [74506, 84191, 95136, 107504, 121479, 137272, 155117, 175282, 198069, 223818, 252914, 285793, 322946, 364929, 412370],
    "Ahorro acumulado ($MXN)": [74506, 158697, 253833, 361337, 482816, 620088, 775205, 950487, 1148556, 1372374, 1625289, 1911082, 2234028, 2598957, 3011327]
}
df = pd.DataFrame(data)

pd.set_option('display.float_format', '{:.4f}'.format)

# Display the DataFrame
print(df)

# Plotting with Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(df['A√±o'], df['Ahorro acumulado ($MXN)'], marker='o', linestyle='-', color='b')
plt.title('Ahorro acumulado a trav√©s de los a√±os')
plt.xlabel('A√±o')
plt.ylabel('Ahorro acumulado ($MXN)')
plt.grid(True)

# Formatear los ejes para evitar la notaci√≥n cient√≠fica
ax = plt.gca()
ax.yaxis.set_major_formatter(ScalarFormatter(useMathText=False))
ax.ticklabel_format(style='plain', axis='y')

plt.show()
```
